Steps involved in EDA are :
* Understanding the data
* cleaning the data
* Analysis of relationship between variables.

**DATA VISUALIZATION**

**COUNT PLOT**

```python
plot=sns.countplot(x="class",data=data)
```

from this visualization we can observe that our data set is a balanced dataset.
in this data set 

0 represents genuine notes

1 represents fake notes 

from the above plot it is clear that genuine notes are more in number than fake notes.

**CORRELATION BETWEEN VARIABLES**

```python
data.corr()
```

**visualizing the correlation using a HEAT MAP**

```python
fig=plt.figure(figsize=(15,15))
sns.heatmap(data.corr(),cmap="Reds",annot=True);
```

This correlation diagram shows that the class column does not have much correlation with other column of the dataset

**PAIR PLOT**

```python
sns.pairplot(data= data,hue= "class")
```

pair plot is showing the bivariate relation between each pair of feautures.

**SCATTER PLOT**

```python

plot=sns.scatterplot(x="skewness",y="variance",hue="class",data=data)
```

From the above visualization, we can observe that genuine notes have more skewness(i.e, symmetric) and more variance than fake notes.

**BOX PLOT**

```python
fig,axs= plt.subplots(figsize=(8,6))
plot=sns.boxplot(x="class",y="variance",data=data)
```

From the above box plot it is clear that the notes having more variance are genuine notes and less variance are fake notes.

**VIOLIN PLOT**

```python
sns.violinplot(x="class",y="curtosis",data=data,size=5)
```

From the above plot we can observe that fake notes have high curtosis compared to genuine notes.we can also observe that fake notes have some duplicate values or outliers.

**JOINT PLOT**

```python
sns.jointplot(x="entropy",y="variance",data=data,height=5)
plt.show()
```

We can also use the seaborn library to make similar plot.A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure.

**HIST PLOT**

```python
data.hist(edgecolor="black",bins=23,figsize=(15,15))
plt.show()
```

**CONCLUSION**
